Using sorted() function

```python
def is_anagram_sorted(str1, str2):
    # Convert to lowercase and remove spaces for a more flexible check
    str1 = str1.replace(" ", "").lower()
    str2 = str2.replace(" ", "").lower()

    # Check if length is the same, if not they can't be anagrams
    if len(str1) != len(str2):
        return False

    # Sort the strings and compare the sorted lists
    return sorted(str1) == sorted(str2)

# Example usage:
word1 = "listen"
word2 = "silent"
print(f"'{word1}' and '{word2}' are anagrams: {is_anagram_sorted(word1, word2)}")

word3 = "hello"
word4 = "world"
print(f"'{word3}' and '{word4}' are anagrams: {is_anagram_sorted(word3, word4)}")
```

2. Using collections.Counter


```python
from collections import Counter

def is_anagram_counter(str1, str2):
    # Convert to lowercase and remove spaces
    str1 = str1.replace(" ", "").lower()
    str2 = str2.replace(" ", "").lower()

    # Counter objects store character counts; direct comparison works
    return Counter(str1) == Counter(str2)

# Example usage:
word1 = "Race"
word2 = "Care"
print(f"'{word1}' and '{word2}' are anagrams: {is_anagram_counter(word1, word2)}")

word3 = "apple"
word4 = "pale"
print(f"'{word3}' and '{word4}' are anagrams: {is_anagram_counter(word3, word4)}")
```

3. Grouping Anagrams from a List

```python
from collections import defaultdict

def group_anagrams(words):
    anagram_groups = defaultdict(list)
    for word in words:
        # Create a unique key by sorting the characters and joining them back to a string
        key = "".join(sorted(word.lower()))
        anagram_groups[key].append(word)
    return list(anagram_groups.values())

# Example usage:
word_list = ["eat", "tea", "tan", "ate", "nat", "bat"]
grouped = group_anagrams(word_list)
print(f"Grouped anagrams: {grouped}")
```
